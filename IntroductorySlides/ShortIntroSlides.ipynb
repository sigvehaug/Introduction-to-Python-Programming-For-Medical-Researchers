## Motivation

Python is a modern programming language which is also much used for Statistical Data Analysis and Machine Learning. It has become a core tool of choice for a wide range of researchers spanning over multiple disciplines. Also medical researcher can profit from all the advanced capabilities and automatisation possibilities not coming with spread sheets.

## Learning objectives

* Participants can/know
* Introduction into programming
* About scripting and Python
* Syntax
* Variables and objects
* Operators
* Built-in data types and structures
