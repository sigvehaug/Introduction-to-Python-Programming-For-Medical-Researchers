```python
import pandas as pd
url = 'https://raw.githubusercontent.com/sigvehaug/Introduction-to-Python-for-Medical-Researchers/master/Data/ISIC_2019_Training_GroundTruth.csv'
df = pd.read_csv(url)
df
```

```python
sums=[df['MEL'].sum(),df['NV'].sum(),df['BCC'].sum(),df['AK'].sum(),df['BKL'].sum(),df['DF'].sum(),df['VASC'].sum(),df['SCC'].sum(),df['UNK'].sum()] 
for r in sums:
  print('%.2f' % (r/df.shape[0]))
```

### Task 2 - Plot maximium 10 or all pictures from each cancer category  

```python
df.columns[1:]
```

```python
from skimage import io
import skimage.filters as skf
import matplotlib.pyplot as plt
url_pre = 'https://raw.githubusercontent.com/sigvehaug/Introduction-to-Python-for-Medical-Researchers/master/Data/'
df_repo = df[:95]      # found that by trial and error

for col in df.columns[1:]:
  dfx = df_repo[df_repo[col]==1]
  print(dfx.shape)
  if dfx.shape[0]>=10:
    nimages = 10
  else:
    nimages = dfx.shape[0]
  print('CANCER TYPE '+col+' '+str(nimages) + ' IMAGES')
  if nimages==0: continue
  plt.figure(figsize=(20,20))
  for i in range(nimages):
    if col=='MEL':   plt.subplot(1,10,i+1)    
    elif col=='NV':  plt.subplot(2,10,i+1)
    #print(i,)
    url = url_pre + dfx.iloc[i,0] + '.jpg'
    try:
      image = io.imread(url)
    except:
      continue  
    #image_h = skf.sobel_v(image[:,:,2])
    plt.imshow(image)
  plt.show()
```
